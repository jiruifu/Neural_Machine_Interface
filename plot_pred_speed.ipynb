```python
import seaborn as sns
import matplotlib.pyplot as plt
import joblib
import pandas as pd
import numpy as np
import os
```

```python
result_path = "result.pkl"
result = joblib.load(result_path)

# Extract data from result
hist_20 = result["prediction_speeds"]["WS20"]["tHist"]
hist_40 = result["prediction_speeds"]["WS40"]["tHist"]
hist_80 = result["prediction_speeds"]["WS80"]["tHist"]
hist_120 = result["prediction_speeds"]["WS120"]["tHist"]
```

```python
# Create a figure with 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()  # Flatten the 2x2 array to make indexing easier

# Plot for window size 40
for i, hist in enumerate(hist_80):
    sns.lineplot(data=hist, ax=axes[i], label=f"WS40-{i}")
    axes[i].set_title(f"Window Size 20 - Trial {i+1}")
    axes[i].set_xlabel("Time Step")
    axes[i].set_ylabel("Speed")

plt.tight_layout()
```
